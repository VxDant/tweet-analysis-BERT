```python
import pandas as pd
from ntscraper import Nitter
from transformers import pipeline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
scraper = Nitter()
```

```python
tweets = scraper.get_tweets('imVkohli',mode='user',number=50)
```

```python
final_tweets = []
for tweet in tweets['tweets']:
    final_tweets.append(tweet['text'])
```

```python
df = pd.DataFrame(final_tweets, columns=['tweet'])
```

```python
emotion_analyzer = pipeline("text-classification", model="bhadresh-savani/bert-base-go-emotion", return_all_scores=True)
```

```python
pip install torch
```

```python

emotion_results = emotion_analyzer(df['tweet'].tolist())

# Extract emotions and scores
emotions = []

for result in emotion_results:
    # Select the emotion with the highest score
    top_emotion = max(result, key=lambda x: x['score'])
    emotions.append(top_emotion['label'])

df['emotion'] = emotions

print(emotions)

emotion_counts = df['emotion'].value_counts()

```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=emotion_counts.index, y=emotion_counts.values, palette='viridis')
plt.title('Recent Tweets from Virat Kohli')
plt.xlabel('Emotion')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
